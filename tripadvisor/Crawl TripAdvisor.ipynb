```python

def getContent():
#     wait = WebDriverWait(driver, 30)
    list_of_content = []
    pageNum = driver.find_element_by_xpath('//*[@id="pager_top2"]')
    pageNum_len = len(pageNum.find_elements_by_xpath("./a"))
#     print(pageNum_len)
    if pageNum_len > 0:
        for no_of_times in range(pageNum_len - 1):
            if no_of_times <= 0 or no_of_times == (pageNum_len - 1):
                xpath = '//html/body/div[4]/div[2]/div[2]/div[2]/div[2]/div[1]/div[8]/div/div/a[' + str(pageNum_len) + ']'
            else:
                xpath = '//html/body/div[4]/div[2]/div[2]/div[2]/div[2]/div[1]/div[8]/div/div/a[' + str(pageNum_len + 1) + ']'
            button = driver.find_element_by_xpath(xpath)
            
            find_comments = driver.find_elements_by_class_name('postBody')
            for content in find_comments:
                p_tags = content.find_elements_by_tag_name('p')
                for p in p_tags:
                    print(p.text)
                    list_of_content.append(p.text)
            button.click()
            driver.implicitly_wait(15)
                        
        find_comments = driver.find_elements_by_class_name('postBody')
        for content in find_comments:
            p_tags = content.find_elements_by_tag_name('p')
            for p in p_tags:
                print(p.text)
                list_of_content.append(p.text)

    else:
        find_comments = driver.find_elements_by_class_name('postBody')
        for content in find_comments:
            p_tags = content.find_elements_by_tag_name('p')
            for p in p_tags:
                print(p.text)
                list_of_content.append(p.text)
                
    return list_of_content
```

```python
def nextPageBtn():
    click_next = driver.find_element_by_xpath('//*[@id="BODY_BLOCK_JQUERY_REFLOW"]/div[2]/div/div[2]/div/div/div/div/div[1]/div/div[1]/div/div[3]/div/div[2]/div/div/div/a[2]')
    click_next.click()
    
#     wait = WebDriverWait(driver, 30)
#     element = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="BODY_BLOCK_JQUERY_REFLOW"]/div[2]/div/div[2]/div/div/div/div/div[1]/div/div[1]/div/div[3]/div/div[1]/div/div[2]/div/div[1]/div/div/div/div[1]/div[1]/div[1]')))
```

```python
def getLastPageNum():
    path = driver.find_element_by_xpath('//*[@id="BODY_BLOCK_JQUERY_REFLOW"]/div[2]/div/div[2]/div/div/div/div/div[1]/div/div[1]/div/div[3]/div/div[2]/div/div/div/div')
    test = path.find_elements_by_xpath('.//a')[-1]
    return test.text
```

```python
def getBoxesTitle():
    driver.implicitly_wait(15)
    content = []
        # whole grid
    path = driver.find_element_by_xpath('//*[@id="BODY_BLOCK_JQUERY_REFLOW"]/div[2]/div/div[2]/div/div/div/div/div[1]/div/div[1]/div/div[3]/div/div[1]/div/div[2]/div/div')
    # Get the title for each boxes in the grid
    test = path.find_elements_by_xpath('.//div/div/div/div[1]/div[1]/div[1]')
    for entry in test:
        if('toilet' in entry.text):
            print(entry.text)
            entry.click()
            openTab()
            content.append(getContent())
            switchBackMainTab()
#             print(entry.text)
    return content
```

```python
def switchBackMainTab():
    driver.close()
    driver.switch_to_window(window_before)
    driver.implicitly_wait(10)

```

```python
def openTab():
    driver.implicitly_wait(10)
    window_after = driver.window_handles[1]
    driver.switch_to_window(window_after)
#     print(driver.page_source)
```
